```python
import matplotlib.pyplot as plt
import numpy as np

c1 = 1
c2 = 1
b1 = 0.4
b2 = 0.2
d1 = 1
d2 = 1

def func1(y1, y2):
    return c1*y1*(1 - b1*y1 - d2*y2)
def func2(y1, y2):
    return -c2*y2*(1 - b2*y2 - d1*y1)

k1 = [0, 0]
k2 = [0, 0]
k3 = [0, 0]
k4 = [0, 0]

t_0 = 0
t_n = 50
h = 0.1
t = np.arange(t_0, t_n, h)
n = len(t)

y1 = np.empty(n)
y1[0] = 2
y2 = np.empty(n)
y2[0] = 2
for i in range(1, n):
    k1[0] = h*func1(y1[i-1], y2[i-1])
    k1[1] = h*func2(y1[i-1], y2[i-1])
    k2[0] = h*func1(y1[i-1]+k1[0]/2, y2[i-1]+k1[1]/2)
    k2[1] = h*func2(y1[i-1]+k1[0]/2, y2[i-1]+k1[1]/2)
    k3[0] = h*func1(y1[i-1]+k2[0]/2, y2[i-1]+k2[1]/2)
    k3[1] = h*func2(y1[i-1]+k2[0]/2, y2[i-1]+k2[1]/2)
    k4[0] = h*func1(y1[i-1]+k3[0], y2[i-1]+k3[1])
    k4[1] = h*func2(y1[i-1]+k3[0], y2[i-1]+k3[1])
    y1[i] = y1[i-1] + (k1[0] + 2*k2[0] + 2*k3[0] + k4[0])/6
    y2[i] = y2[i-1] + (k1[1] + 2*k2[1] + 2*k3[1] + k4[1])/6

#create the graph
plt.plot(t, y1, marker="o", color = "red", linestyle = "--")
plt.plot(t, y2, marker="v", color = "blue", linestyle = ":")
plt.xlabel('t')
plt.ylabel('y')
plt.legend(['y1','y2'])
plt.show()
```
